```python
import tkinter as tk
from tkinter import ttk, messagebox
from PIL import Image, ImageTk
import pandas as pd
import plotly.express as px
import webbrowser
import folium
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap

# Load datasets (mock placeholders, replace with actual file paths)
df_kpk = pd.read_csv(r"C:/Users/HP/OneDrive/Desktop/prog for ai/Project solo semester/kpk data_cleaned/crime_rate_kpk.csv")
df_punjab = pd.read_csv(r"C:/Users/HP/OneDrive/Desktop/prog for ai/Project solo semester/punjab data_cleaned/combined_data.csv")
df_pak1=pd.read_csv(r"C:/Users/HP/OneDrive/Desktop/prog for ai/Project solo semester/pakistan overall_cleaned/all number of crime.csv")
df_pak2=pd.read_csv(r"C:/Users/HP/OneDrive/Desktop/prog for ai/Project solo semester/pakistan overall_cleaned/total number of reported crime.csv")
data_cityWise=pd.read_csv(r"C:/Users/HP/OneDrive/Desktop/prog for ai/Project solo semester/pakistan overall_cleaned/overall_province_wise_clean_crime_data.csv")

# Initialize main Tkinter application
root = tk.Tk()
root.title("Crime Analysis Project")
root.geometry("1600x800")


#Load background image
bg_image = Image.open("istockphoto-1404189261-612x612.jpg")  
bg_image = bg_image.resize((1600, 800))  
bg_photo = ImageTk.PhotoImage(bg_image)

# Create a background label to hold the image
def create_background(frame):
    if not hasattr(frame, "_bg_applied"):  # Check if background is already applied
        bg_label = tk.Label(frame, image=bg_photo)
        bg_label.place(relwidth=1, relheight=1)  # Make the label fill the entire frame
        bg_label.image = bg_photo  # Keep a reference to prevent garbage collection
        frame._bg_applied = True  # Mark background as applied


# Create a container for all frames
frames = {}

#Create a background label to hold the image
def create_background(frame):
    bg_label = tk.Label(frame, image=bg_photo)
    bg_label.place(relwidth=1, relheight=1)  # Make the label fill the entire frame
    bg_label.image = bg_photo  # Keep a reference to the image to prevent it from being garbage collected


def show_frame(frame_name):
    """
    Show the specified frame by name, hiding all others.
    """
    for frame in frames.values():
        frame.pack_forget()
    frames[frame_name].pack(fill="both", expand=True)


# Main Menu Frame
def create_main_menu():
    frame = tk.Frame(root)
    frames["main_menu"] = frame
    create_background(frame)
    tk.Label(frame, text=" CRIME ANALYSIS ", font=("Arial", 50, "bold", "italic"), fg="black").place(relx=0.5, rely=0.17, anchor="center")
    
    tk.Label(frame, 
        text=(' "Hey UserðŸ‘‹! \n This is Crime analysis Dashboard : \nCrime Data Insights at Your Fingertips\n Explore and analyze crime trends across Pakistan with interactive visualizations.\n Dive into district-wise, province-level, and nationwide data to uncover patterns \n also make informed decisions effortlessly! "'),
        font=("Tegami Type", 20,'bold'), fg="black", justify="center").place(relx=0.5, rely=0.4, anchor="center")
    
    tk.Button(frame, text="START ANALYSIS", font=("Tegami Type", 12, "bold"), command=lambda: show_frame("analysis_menu"),width=30, height=3, bg="green", fg="white").place(relx=0.5, rely=0.6, anchor="center")
    tk.Label(frame,text="BY: MUNAZZAH \n PROGRAMMING FOR AI PROJECT",font=("Tegami Type", 14,'bold'), fg="black", justify="left").place(relx=0.9, rely=0.9, anchor="e")
    return frame

# Analysis Menu Frame
def create_analysis_menu():
    frame = tk.Frame(root)
    frames["analysis_menu"] = frame
    create_background(frame)
    tk.Label(frame, text=" ANALYSIS MENU ", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.12, anchor="center")
    
    tk.Button(frame, text="PAKISTAN OVERALL ANALYSIS", command=lambda: show_frame("pakistan_analysis"), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.3, rely=0.3, anchor="center")
    
    tk.Button(frame, text="PROVINCE WISE ANALYSIS ", command=lambda: show_frame("citywise_analysis"), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.7, rely=0.3, anchor="center")
    
    tk.Button(frame, text="KPK CRIME ANALYSIS", command=lambda: show_frame("kpk_analysis"), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.3, rely=0.6, anchor="center")
    
    tk.Button(frame, text="PUNJAB CRIME ANALYSIS ", command=lambda: show_frame("punjab_analysis_page"), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.7, rely=0.6, anchor="center")
    
    tk.Button(frame, text="BACK", command=lambda: show_frame("main_menu"), width=20, height=2, bg="black", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.5, rely=0.9, anchor="center")
    return frame

# Pakistan Overall Analysis Frame
def create_pakistan_analysis():

    frame = tk.Frame(root)
    frames["pakistan_analysis"] = frame
    create_background(frame)
    tk.Label(frame, text="Pakistan Overall Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    tk.Label(frame,text=' "Data speaks louder than opinions\n uncover the truth hidden in the numbers! " ',font=("Arial", 20, "bold", "italic")).place(relx=0.5, rely=0.23, anchor="center" )
    
    tk.Button(frame, text=" VISUALIZE ALL REPORTED CRIME DATA ", command=lambda: show_frame("visualize_all_reported"), width=33, height=3, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.5, rely=0.4, anchor="center")
    
    tk.Button(frame, text=" VISUALIZE CRIME TRENDS BY YEAR ",  command=lambda: show_frame("visualize_crime_type"), width=33, height=3, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.5, rely=0.6, anchor="center")

    tk.Button(frame, text="BACK",  command=lambda: show_frame("analysis_menu"),  width=20, height=2, bg="black", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.5, rely=0.8, anchor="center")
    return frame


# pakistan all reported analysis frame
def create_visualize_all_reported():

    frame = tk.Frame(root)
    frames["visualize_all_reported"] = frame
    create_background(frame)
    tk.Label(frame, text="All Reported Crimes Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    tk.Label(frame, text="Visualize all Reported Crimes in Pakistan (2010-2020).",  font=("Arial", 20,'bold'), justify="center").place(relx=0.5, rely=0.3, anchor="center")
    
    tk.Button(frame, text=" LETS VISUALIZE ",  command=lambda: visualize_all_reportedCrimes(df_pak2), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15,'bold')).place(relx=0.5, rely=0.5, anchor="center")
   
    tk.Button(frame, text=" BACK ",  command=lambda: show_frame("pakistan_analysis"),  width=20, height=2, bg="black", fg="white", font=("Tegami Type", 10)).place(relx=0.5, rely=0.7, anchor="center")
    return frame

#visualize all reported crime
def visualize_all_reportedCrimes(df):
    # Preprocess the Total Crime column
    df['Total Crime'] = df['Total Crime'].astype(str).str.replace(',', '').astype(int)

    fig = px.bar(df, x='year', y='Total Crime', title="All Reported Crimes Over the Years (2010-2020)", labels={'All Reported Crimes': 'Number of All Reported Crimes'},color='year')
    
    # Save the plot as an HTML file

    fig.write_html("all_reported_crimes_analysis.html")
    
    webbrowser.open("all_reported_crimes_analysis.html")

#All crime trends over years
def create_visualize_crime_type():

    frame = tk.Frame(root)
    frames["visualize_crime_type"] = frame
    create_background(frame)
    tk.Label(frame, text=" Crime Type Analysis By Year ", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    
    tk.Label(frame, text=' "Uncover crime trends year by year!\n Input a year, choose your graph style, and bring data insights to life.\n Transform numbers into stories that matter." ', font=("Arial", 15,'bold'),  justify="center").place(relx=0.5,rely=0.24, anchor="center")
    
    tk.Label(frame, text="SELECT YEAR :", font=("Arial", 15,'bold')).place(relx=0.4, rely=0.35, anchor="center")
    year_combobox = ttk.Combobox(frame, values=[ 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020], width=20, font=("Arial", 12,'bold'), justify="center",state='readonly')
    year_combobox.place(relx=0.6, rely=0.35, anchor="center")
    year_combobox.set(2020)  # Default selection

    tk.Label(frame, text="SELECT GRAPH TYPE :", font=("Arial", 15,'bold')).place(relx=0.4, rely=0.5, anchor="center")
    graph_type_combobox = ttk.Combobox(frame, values=["bar", "pie"], width=20, font=("Arial", 12,'bold'),state='readonly')
    graph_type_combobox.place(relx=0.6, rely=0.5, anchor="center")
    graph_type_combobox.set("bar")  # Default selection

    def on_submit():
        year = year_combobox.get().strip()
        graph_type = graph_type_combobox.get().strip()

        if not year.isdigit() or not (2010 <= int(year) <= 2020):
            messagebox.showerror("Input Error", "Please enter a valid year between 2010 and 2020.")
            return

        visualize_pak_crimetype(df_pak1, int(year), graph_type)
    
    
    tk.Button(frame, text="LETS VISUALIZE",   command=on_submit, width=30, height=3, bg="green", fg="white", font=("Tegami Type", 10,'bold')).place(relx=0.5, rely=0.7, anchor="center")

    
    tk.Button(frame, text=" BACK ",  command=lambda: show_frame("pakistan_analysis"),  width=30, height=2, bg="black", fg="white", font=("Tegami Type", 10,'bold')).place(relx=0.5, rely=0.8, anchor="center")
    return frame

#visualize
def visualize_pak_crimetype(df, year, graph_type='bar'):

    filtered_df = df[df['year'] == year]
    
    # Sort by Crime Count to ensure proper ordering in the visualization
    filtered_df = filtered_df.sort_values(by='Crime Count', ascending=False)
    
   
    if graph_type == "bar":
        fig = px.bar(filtered_df, x='Crime Type', y='Crime Count', color='Crime Type', title=f"Crime Trends in {year}", labels={'Crime Count': 'Number of Crimes'})
    elif graph_type == "pie":
        fig = px.pie(filtered_df, values='Crime Count', names='Crime Type', title=f"Crime Distribution in {year}", color_discrete_sequence=px.colors.sequential.RdBu)
    else:
        messagebox.showwarning("Invalid graph type. Choose from 'bar', 'line', or 'scatter'.")
    

    fig.write_html("deatil_by_year_analysis.html")
    
    webbrowser.open("deatil_by_year_analysis.html")    

#manually added latitude and longitude for heap map
district_coords = {
    "Punjab": {"Latitude": 31.5497, "Longitude": 74.3436},  # Lahore (capital city of Punjab)
    "Sindh": {"Latitude": 24.8607, "Longitude": 67.0011},   # Karachi (capital city of Sindh)
    "KPK": {"Latitude": 34.0150, "Longitude": 71.5249},     # Peshawar (capital city of KPK)
    "Balochistan": {"Latitude": 30.1798, "Longitude": 66.9750}, # Quetta (capital city of Balochistan)
    "Islamabad": {"Latitude": 33.6844, "Longitude": 73.0479}, # Capital of Pakistan
    "Railways": {"Latitude": 33.6844, "Longitude": 73.0479},  # Same as Islamabad (Assumed)
    "G.B": {"Latitude": 35.3, "Longitude": 75.5833},         # Gilgit-Baltistan (Assumed)
    }
   
def add_coordinates_to_data(df):
    
    for province in df['district'].unique():
        # Manually get coordinates from the dictionary
        coordinates = district_coords.get(province)
        if coordinates:
            df.loc[df['district'] == province, 'Latitude'] = coordinates['Latitude']
            df.loc[df['district'] == province, 'Longitude'] = coordinates['Longitude']
        else:
            print(f"Coordinates for {province} not found.")  # Debugging log

    return df

# Add coordinates
df_with_coords = add_coordinates_to_data(data_cityWise)

#pakistan province visualization
def create_pakistan_citywise_analysis():
    frame = tk.Frame(root)
    frames["citywise_analysis"] = frame
    create_background(frame)
    HIGH_ALERT_THRESHOLD = 500  # Define the high alert threshold for total crimes

    def on_submit():
        # Get selected provinces from the listbox
        selected_provinces_indices = province_listbox.curselection()
        selected_provinces = [province_listbox.get(i) for i in selected_provinces_indices]
        
        # Ensure that at least one province is selected
        if not selected_provinces:
            messagebox.showwarning("No provinces selected", "Please select at least one province.")
            return
        
        selected_year = int(year_combobox.get().strip())
        selected_graph = graph_combobox.get().strip()
        
        visualize_citywise_data(data_cityWise, selected_provinces, selected_year, selected_graph)

    tk.Label(frame, text="PROVINCE/CITY CRIME ANALYSIS", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    tk.Label(frame, text="SELECT YEAR:", font=("Arial", 15, 'bold')).place(relx=0.2, rely=0.4, anchor="w")
    year_combobox = ttk.Combobox(frame, values=[2018, 2019, 2020], width=20, font=("Arial", 12, 'bold'), state='readonly', justify="center")
    year_combobox.set(2020)
    year_combobox.place(relx=0.5, rely=0.4, anchor="center")

    tk.Label(frame, text="SELECT GRAPH:", font=("Arial", 15, 'bold')).place(relx=0.2, rely=0.5, anchor="w")
    graph_combobox = ttk.Combobox(frame, values=['bar','pie' ,'map'], width=20, font=("Arial", 12, 'bold'), state='readonly', justify="center")
    graph_combobox.place(relx=0.5, rely=0.5, anchor="center")

    tk.Label(frame, text="SELECT PROVINCE:", font=("Arial", 15, 'bold')).place(relx=0.2, rely=0.65, anchor="w")
    province_listbox = tk.Listbox(frame, selectmode=tk.MULTIPLE, font=("Arial", 12, 'bold'), height=6, width=20)
    provinces = ["Punjab", "Sindh", "KPK", "Balochistan", "Islamabad", "Railways", "G.B"]
    for province in provinces:
        province_listbox.insert(tk.END, province)
    province_listbox.place(relx=0.5, rely=0.65, anchor="center")

    
    tk.Button(frame, text=" VISUALIZE ", command=on_submit, width=30, height=3, bg="green", fg="white", font=("Tegami Type", 10, 'bold')).place(relx=0.5, rely=0.8, anchor="center")
    tk.Button(frame, text=" BACK ", command=lambda: show_frame("analysis_menu"), width=20, height=2, bg="black", fg="white", font=("Tegami Type", 10, 'bold')).place(relx=0.5, rely=0.9, anchor="center")

    return frame

#visualize province -wise data
def visualize_citywise_data(df, provinces, year, graph_type, top_bottom=0):
    # Simplify the filtering process
    filtered_data = df[df['district'].isin(provinces) & (df['year'] == year)]

    if filtered_data.empty:
        messagebox.showwarning("No data available for selected filters")
        return
    

    # Aggregate data for visualization
    aggregated_data = filtered_data.groupby(['district', 'Crime Type', 'year'])['Total Crime'].sum().reset_index()
    
    # **High Alert Logic**: Check for districts exceeding the threshold
    high_alert_districts = aggregated_data[aggregated_data['Total Crime'] > 500]

    
    # Handle map visualization using Folium
    if graph_type == "map":
        visualize_map(filtered_data)  # Call map function
    elif graph_type == "bar":
        fig = px.bar(aggregated_data, x='district', y='Total Crime', title="City wise Crime Analysis", labels={"district": "City", "Total Crime": "Number of Incidents", "Crime Type": "Type of Crime"},barmode='group', color='Crime Type')
        # Save the plot as an HTML file and open it
        fig.write_html('pakistan_citywise_crimes.html')
        webbrowser.open("pakistan_citywise_crimes.html")
    elif graph_type=="pie":
        fig_pie = px.pie(aggregated_data,values='Total Crime',names='Crime Type',title="Crime Type Distribution ",color_discrete_sequence=px.colors.sequential.RdBu)
        # Save the plot as an HTML file and open it
        fig_pie.write_html('pakistan_citywise_crimes.html')
        webbrowser.open("pakistan_citywise_crimes.html")
    else:
        messagebox.showwarning("Invalid graph type selected. Choose from 'bar' or 'map'.")
    

    # Show high alert notifications
    if not high_alert_districts.empty:
        alert_message = (
            "High Alert Districts:\n" +
            "\n".join(
                f"\n{row['district']}:{row['Crime Type']}: {row['Total Crime']} crimes \n"
                for _, row in high_alert_districts.iterrows()
            )
        )
        messagebox.showwarning("High Alert!", alert_message)
    else:
        messagebox.showinfo("All Clear", "No districts require high alert.")



#visualize using folium and heap map
def visualize_map(df):
    # Ensure 'Latitude' and 'Longitude' are present
    if 'Latitude' not in df.columns or 'Longitude' not in df.columns:
        messagebox.showwarning("Missing geographic data", "Latitude and Longitude are required for map visualization.")
        return

    # Create the map centered on Pakistan
    m = folium.Map(location=[30.3753, 69.3451], zoom_start=5)
    # # Initialize a MarkerCluster to handle multiple markers
    marker_cluster = MarkerCluster().add_to(m)
    

    for _,row in df.iterrows():
        folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"{row['district']} - {row['Total Crime']} crimes ({row['Crime Type']} in {row['year']})",
        tooltip=f"Click for details on {row['district']}",
        radius=10,
        color="red",
        fill=True,
        fill_color="red",
        fill_opacity=0.6,
    ).add_to(marker_cluster)
        
   # Prepare data for heatmap: 
    heatmap_data = df[['Latitude', 'Longitude', 'Total Crime']].values.tolist()
    # Add the heatmap layer
    HeatMap(heatmap_data, radius=30, blur=10, max_zoom=20).add_to(m)


    # Save the map as an HTML file and open it
    m.save('pakistan_citywise_crimes_map.html')
    webbrowser.open("pakistan_citywise_crimes_map.html")

    
# KPK Analysis Frame
def create_kpk_analysis():
    frame = tk.Frame(root)
    frames["kpk_analysis"] = frame
    create_background(frame)

    tk.Label(frame, text="KPK Crime Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    tk.Label(frame, text=' "Insights into KPKs crime data with interactive district and crime type analysis."', font=("Arial", 15,'bold','italic')).place(relx=0.5,rely=0.2, anchor="center")
    
    tk.Button(frame, text="KPK DISTRICT ANALYSIS", command=lambda: show_frame("kpk_district_analysis"), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15, 'bold')).place(relx=0.5, rely=0.4, anchor="center")

    tk.Button(frame, text="KPK CRIME TREND ANALYSIS", command=lambda: show_frame("kpk_crime_type_analysis"), width=30, height=2, bg="green", fg="white", font=("Tegami Type", 15, 'bold')).place(relx=0.5, rely=0.6, anchor="center")



    tk.Button(frame, text="BACK", command=lambda: show_frame("analysis_menu"), width=30, height=2, bg="black", fg="white").place(relx=0.5, rely=0.75, anchor="center")
    
    
    return frame

# Function to create the KPK district analysis frame
def create_kpk_district_analysis():
    frame = tk.Frame(root)
    frames["kpk_district_analysis"] = frame
    create_background(frame)
    

    # Ensure the 'year' column is a string type
    df_kpk['year'] = df_kpk['year'].astype(str)
    
    # Function to update the available years based on district selection
    def update_years(event=None):
        district = district_var.get().strip().lower()  # Get the selected district
        
        if district:  # If a district is selected
            available_years = sorted(df_kpk[df_kpk['district'].str.lower() == district]['year'].unique())
        else:
            available_years = sorted(df_kpk['year'].unique())
        
        # Update the year dropdown values only if the district has changed
        year_dropdown['values'] = available_years
        if year_var.get() not in available_years:  # If current year no longer available, reset to first available year
            year_var.set(available_years[-1] if available_years else '')  # Default to last
    
    # Function to handle the submission of inputs
    def submit_inputs():
        district = district_var.get().strip().lower() or None
        year = year_var.get().strip() or None
   
        if not district:
            messagebox.showwarning("Invalid Input", "Please select a district.")
            return  
        
        
        if not year:
            messagebox.showwarning("Invalid Input", "Please select a year.")
            return 
        
        # Ensure that 'district' is in a list format
        provinces = [district]  # Wrap the district string in a list
        
    # Ensure year is a string
        year = str(year)

    # Normalize the 'district' column in the DataFrame to lowercase
        df_kpk['district'] = df_kpk['district'].str.lower()

    # Filter data based on district and year
        filtered_data = df_kpk[df_kpk['district'].isin(provinces) & (df_kpk['year'] == year)]

    # Check if filtered data exists
        if filtered_data.empty:
            messagebox.showwarning("No Data", "No data found for the selected district and year.")
            return
        # Get the selected graph type
        graph_type = graph_type_var.get().strip().lower()

        if graph_type == 'bar':
            
            visualize_citywise_data(df_kpk, provinces, year, graph_type='bar')
        elif graph_type == 'pie':
            visualize_citywise_data(df_kpk, provinces, year, graph_type='pie')

  
    # Create input UI components
        
    tk.Label(frame, text="KPK District Crime Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    tk.Label(frame, text="SELECT GRAPH:", font=("Arial", 20, "bold", 'italic')).place(relx=0.4, rely=0.3, anchor="center")
    graph_type_var = tk.StringVar()
    graph_type_dropdown = ttk.Combobox(frame, textvariable=graph_type_var, values=['bar','pie'], state='readonly', font=("Arial", 14, "bold"))
    graph_type_dropdown.place(relx=0.4, rely=0.4, anchor="center")

    # Create District Combobox (dynamically filled from dataset)
    tk.Label(frame, text="Select District:", font=("Arial", 20, "bold", 'italic')).place(relx=0.6, rely=0.3, anchor="center")
    district_var = tk.StringVar()
    districts = df_kpk['district'].unique().tolist()  # Get unique districts from the dataset
    district_dropdown = ttk.Combobox(frame, textvariable=district_var, values=districts, state='readonly', font=("Arial", 14, "bold"))
    district_dropdown.bind("<<ComboboxSelected>>", update_years)  # Update years when district changes
    district_dropdown.place(relx=0.6, rely=0.4, anchor="center")

    # Create Year Combobox (dynamically filled based on district)
    tk.Label(frame, text="Select Year:", font=("Arial", 20, "bold", 'italic')).place(relx=0.5, rely=0.5, anchor="center")
    year_var = tk.StringVar()
    year_dropdown = ttk.Combobox(frame, textvariable=year_var, state='readonly', font=("Arial", 14, "bold"))
    year_dropdown.place(relx=0.5, rely=0.55, anchor="center")

    # Initial population of years (if the first district is selected)
    update_years()

    
    submit_button = tk.Button(frame, text="LET'S VISUALIZE !!", command=submit_inputs, bg='green', fg='white', width=30, height=3, font=("Tegami Type", 10, "bold"))
    submit_button.place(relx=0.5, rely=0.65, anchor="center")

    # Back button to return to KPK analysis menu
    tk.Button(frame, text=" BACK ", command=lambda: show_frame("kpk_analysis"), width=20, height=2, bg="black", fg="white", font=("Tegami Type", 10)).place(relx=0.5, rely=0.75, anchor="center")

    return frame

# Function to create the KPK Crime Type Analysis frame
def create_kpk_crime_type_analysis():
    frame = tk.Frame(root)
    frames["kpk_crime_type_analysis"] = frame
    frame.pack(fill="both", expand=True)
    create_background(frame)

    HIGH_ALERT_THRESHOLD = 500  # Define the high alert threshold for total crimes

  
    # Function to update available years based on Crime Type selection
    def update_years(event=None):
        crime_type = crime_type_var.get().strip().lower()  # Get the selected Crime Type
        
        if crime_type:  # If a Crime Type is selected
            available_years = sorted(df_kpk[df_kpk['Crime Type'].str.lower() == crime_type]['year'].unique())
        else:
            available_years = sorted(df_kpk['year'].unique())
        year_dropdown['values'] = available_years
        if year_var.get() not in available_years:  # current year no longer available, reset to first year
            year_var.set(available_years[-1] if available_years else '')  # default to last

    # Function to handle the submission of inputs and graph generation
    def submit_inputs():
        crime_type = crime_type_var.get().strip().lower()
        year = year_var.get().strip() or None
        top_bottom = top_bottom_var.get().strip().lower()
        number = number_var.get()
  
        if not crime_type:
            messagebox.showwarning("Invalid Input", "Please select a crime type.")
            return

        if not year:
            messagebox.showwarning("Invalid Input", "Please select a year.")
            return

        if not number:
            messagebox.showwarning("Invalid Input", "Please select a number for top/bottom.")
            return
        
        # Filter the data by the selected crime type and year
        filtered_data = df_kpk[df_kpk['Crime Type'].str.lower() == crime_type]
        if year:
            filtered_data = filtered_data[filtered_data['year'] == year]
        
        # Sort data based on total crime and select top/bottom N
        sorted_data = filtered_data.groupby(['district', 'Crime Type', 'year'])['Total Crime'].sum().reset_index()
        sorted_data = sorted_data.sort_values(by="Total Crime", ascending=True)

           # Select top/bottom N rows
        if top_bottom == 'top':
        # For "top", sort by descending Total Crime and select top N
            sorted_data = sorted_data.sort_values(by="Total Crime", ascending=False)
            top_bottom_n = sorted_data.head(number)
        elif top_bottom == 'bottom':
        # For "bottom", sort by ascending Total Crime and select top N
            sorted_data = sorted_data.sort_values(by="Total Crime", ascending=True)
            top_bottom_n = sorted_data.head(number)

        # **High Alert Logic**: Check for districts exceeding the threshold
        high_alert_districts = sorted_data[sorted_data['Total Crime'] > HIGH_ALERT_THRESHOLD]

        if not top_bottom_n.empty:
            visualize_crime_type_data(top_bottom_n)

        # Show high alert notifications
        if not high_alert_districts.empty:
            alert_message = (
                "High Alert Districts:\n" +
                "\n".join(
                    f"{row['district']}: {row['Total Crime']} crimes \n"
                    for _, row in high_alert_districts.iterrows()
                )
            )
            messagebox.showwarning("High Alert!", alert_message)
        else:
            messagebox.showinfo("All Clear", "No districts require high alert.")
        
  
    # Function to visualize the crime type data (graph generation)
    def visualize_crime_type_data(data):
        
        
        # Create a Plotly bar plot for Crime Type Analysis
        fig = px.bar(data, x='district', y='Total Crime', color='district', title="Crime Type Analysis",labels={"district": "District", "Total Crime": "Number of Incidents", "Crime Type": "Type of Crime"})
        
        fig.write_html("crime_type_analysis.html")
        webbrowser.open("crime_type_analysis.html")   

    #GUI 
        
    tk.Label(frame, text="KPK Crime Type Analysis", font=("Arial", 30, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    tk.Label(frame, text=" SELECT CRIME : ", font=("Arial", 22, "bold", 'italic')).place(relx=0.4, rely=0.3, anchor="center")
    crime_type_var = tk.StringVar()
    crime_types = df_kpk['Crime Type'].unique().tolist()  # Get unique Crime Types from the dataset
    crime_type_dropdown = ttk.Combobox(frame, textvariable=crime_type_var, values=crime_types, state='readonly', font=("Arial", 15, "bold"))
    crime_type_dropdown.bind("<<ComboboxSelected>>", update_years)  # Update years when crime type changes
    crime_type_dropdown.place(relx=0.4, rely=0.4, anchor="center")

    # Creating Year Combobox (dynamically filled based on Crime Type)
    tk.Label(frame, text=" Select Year: ", font=("Arial", 22, "bold", 'italic')).place(relx=0.6, rely=0.3, anchor="center")
    year_var = tk.StringVar(value=2020)
    year_dropdown = ttk.Combobox(frame, textvariable=year_var, state='readonly', font=("Arial", 14, "bold"))
    year_dropdown.place(relx=0.6, rely=0.4, anchor="center")

    # Creating Top/Bottom Combobox
    tk.Label(frame, text="Top/Bottom:", font=("Arial", 20, "bold", 'italic')).place(relx=0.4, rely=0.6, anchor="center")
    top_bottom_var = tk.StringVar(value='top')
    top_bottom_dropdown = ttk.Combobox(frame, textvariable=top_bottom_var, values=['top', 'bottom'], state='readonly', font=("Arial", 14, "bold"))
    top_bottom_dropdown.place(relx=0.4, rely=0.7, anchor="center")

    # Creating Number Combobox (1-7)
    tk.Label(frame, text="Number (1-7):", font=("Arial", 20, "bold", 'italic')).place(relx=0.6, rely=0.6, anchor="center")
    number_var = tk.IntVar(value='5')
    number_dropdown = ttk.Combobox(frame, textvariable=number_var, values=[i for i in range(1, 8)], state='readonly', font=("Arial", 14, "bold"))
    number_dropdown.place(relx=0.6, rely=0.7, anchor="center")

    # Submit button
    submit_button = tk.Button(frame, text="LET'S VISUALIZE !!", command=submit_inputs, bg='green', fg='white', width=30, height=3, font=("Tegami Type", 10, "bold"))
    submit_button.place(relx=0.5, rely=0.8, anchor="center")


    # Back button to return to KPK analysis menu
    tk.Button(frame, text=" BACK ", command=lambda: show_frame("kpk_analysis"), width=30, height=2, bg="black", fg="white", font=("Tegami Type", 10,'bold')).place(relx=0.5, rely=0.9, anchor="center")

    return frame

# Punjab Analysis Page Frame
def punjab_analysis_page():
    frame = tk.Frame(root)
    frames["punjab_analysis_page"] = frame
    create_background(frame)
    tk.Label(frame, text="Punjab Crime Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    # Button for Crime by Type
    tk.Button(
        frame, text="PUNJAB CRIME TREND ANALYSIS", bg="green", fg="white", width=30, height=3, font=("Arial", 14, "bold"),
        command=lambda: show_frame("punjab_crime_type_analysis")
    ).place(relx=0.5, rely=0.4, anchor="center")

    # Button for Total Crime in District
    tk.Button(
        frame, text="PUNJAB DISTRICT ANALYSIS", bg="green", fg="white", width=30, height=3, font=("Arial", 14, "bold"),
        command=lambda: show_frame("punjab_district_crime_analysis")
    ).place(relx=0.5, rely=0.6, anchor="center")

    # Back Button
    tk.Button(
        frame, text="BACK", command=lambda: show_frame("create_analysis_menu"), width=30, height=2,
        bg="black", fg="white", font=("Tegami Type", 10, "bold")
    ).place(relx=0.5, rely=0.8, anchor="center")

    return frame

# Total Crime in District Function
def punjab_district_crime_analysis():
    frame = tk.Frame(root)
    frames["punjab_district_crime_analysis"] = frame
    create_background(frame)
    tk.Label(frame, text="Punjab Total Crime Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    
    # Select Graph Type
    tk.Label(frame, text="Select Graph Type:", font=("Arial", 20, "bold")).place(relx=0.3, rely=0.3, anchor="center")
    graph_type_var = tk.StringVar(value="Bar")
    graph_type_dropdown = ttk.Combobox(frame, textvariable=graph_type_var, values=["Bar", "Pie"], state="readonly", font=("Arial", 14, "bold"))
    graph_type_dropdown.place(relx=0.3, rely=0.4, anchor="center")

    # Select District (only one district using Combobox)
    tk.Label(frame, text="Select District:", font=("Arial", 20, "bold")).place(relx=0.6, rely=0.3, anchor="center")
    district_var = tk.StringVar()
    districts = df_punjab['district'].unique().tolist()
    district_combobox = ttk.Combobox(frame, textvariable=district_var, values=districts, state="readonly", font=("Arial", 14, "bold"))
    district_combobox.place(relx=0.6, rely=0.4, anchor="center")

    # Year Dropdown
    tk.Label(frame, text="Select Year:", font=("Arial", 20, "bold")).place(relx=0.3, rely=0.5, anchor="center")
    year_var = tk.StringVar()
    year_dropdown = ttk.Combobox(frame, textvariable=year_var, state="readonly", font=("Arial", 14, "bold"))
    year_dropdown.set(2020)
    year_dropdown.place(relx=0.3, rely=0.6, anchor="center")

    def update_years():
        selected_district = district_combobox.get()  # Only one district selected
        if selected_district:
            available_years = sorted(df_punjab[df_punjab['district'] == selected_district]['year'].unique())
        else:
            available_years = sorted(df_punjab['year'].unique())
        year_dropdown['values'] = available_years
        if year_var.get() not in available_years:  # current year no longer available, reset to first year
            year_var.set(available_years[-1] if available_years else '')  # default to last if available

    # Update years whenever district is selected
    district_combobox.bind("<<ComboboxSelected>>", lambda event: update_years())

    # Submit Button
    def submit_inputs():
        selected_graph = graph_type_var.get()
        selected_district = district_combobox.get()  # Only one district selected
        selected_year = year_var.get()

        if not selected_district:
            messagebox.showwarning("Invalid Input", "Please select a district.")
            return
        if not selected_year:
            messagebox.showwarning("Invalid Input", "Please select a year.")
            return
        
        # Filter data for the selected district and year
        filtered_data = df_punjab[(df_punjab['district'] == selected_district) & (df_punjab['year'] == int(selected_year))]
        if filtered_data.empty:
            messagebox.showinfo("No Data", "No data available for the selected criteria.")
            return
        
        # Generate the graph
        if selected_graph == "Bar":
            fig = px.bar(filtered_data, x="Crime Type", y="Total Crime",color='Crime Type',title=f"Total Crime Analysis - {selected_year}", labels={"Crime Type": "Crime Trends", "Total Crime": "Number of Crimes"})
        elif selected_graph == "Pie":
            fig = px.pie(filtered_data, names="Crime Type", values="Total Crime",title=f"Total Crime Distribution - {selected_year}",color_discrete_sequence=px.colors.sequential.RdBu)
        
        # Save and open the graph
        fig.write_html("punjab_total_crime_analysis.html")
        webbrowser.open("punjab_total_crime_analysis.html")

    tk.Button(
        frame, text="SHOW GRAPH", command=submit_inputs, bg="green", fg="white", width=20, height=2, font=("Arial", 14, "bold")
    ).place(relx=0.5, rely=0.75, anchor="center")
    
    # Back Button
    tk.Button(
        frame, text="BACK", command=lambda: show_frame("punjab_analysis_page"), width=20, height=2, bg="black", fg="white", font=("Arial", 10, "bold")).place(relx=0.5, rely=0.85, anchor="center")

    return frame

# Punjab Analysis Frame
def punjab_crime_type_analysis():
    frame = tk.Frame(root)

    frames["punjab_crime_type_analysis"] = frame
    create_background(frame)

    HIGH_ALERT_THRESHOLD = 500  # Define the high alert threshold for total crimes

    # Function to update available years based on Crime Type selection
    def update_years(event=None):
        crime_type = crime_type_var.get().strip().lower()  # Get the selected Crime Type
        if crime_type:  # If a Crime Type is selected
            available_years = sorted(df_punjab[df_punjab['Crime Type'].str.lower() == crime_type]['year'].unique())
        else:
            available_years = sorted(df_punjab['year'].unique())
        year_dropdown['values'] = available_years
        if year_var.get() not in available_years:  # current year no longer available, reset to first year
            year_var.set(available_years[-1] if available_years else '')  # default to last

    # Function to handle the submission of inputs and graph generation
    def submit_inputs():
        crime_type = crime_type_var.get().strip().lower()
        year = year_var.get().strip() or None
        top_bottom = top_bottom_var.get().strip().lower()
        number = number_var.get()
        
        if not crime_type:
            messagebox.showwarning("Invalid Input", "Please select a crime type.")
            return

        if not year:
            messagebox.showwarning("Invalid Input", "Please select a year.")
            return

        if not number:
            messagebox.showwarning("Invalid Input", "Please select a number for top/bottom.")
            return
        
   
        # Filter by Crime Type and Year
        filtered_data = df_punjab[df_punjab['Crime Type'].str.lower() == crime_type]
        if year:
            filtered_data = filtered_data[filtered_data['year'] == int(year)]  # Ensure the year is an integer

            # Exclude Punjab district from the data
        
        filtered_data = filtered_data[filtered_data['district'] != 'PUNJAB']


    # Sort data and select top/bottom N rows
        sorted_data = filtered_data.groupby(['district', 'Crime Type', 'year'])['Total Crime'].sum().reset_index()
        sorted_data = sorted_data.sort_values(by="Total Crime", ascending=(True))

    # Select top/bottom N rows
        if top_bottom == 'top':
        # For "top", sort by descending Total Crime and select top N
            sorted_data = sorted_data.sort_values(by="Total Cri-me", ascending=False)
            top_bottom_n = sorted_data.head(number)
        elif top_bottom == 'bottom':
        # For "bottom", sort by ascending Total Crime and select top N
            sorted_data = sorted_data.sort_values(by="Total Crime", ascending=True)
            top_bottom_n = sorted_data.head(number)

        # Sort data based on total crime and select top/bottom N
        sorted_data = filtered_data.groupby(['district', 'Crime Type', 'year'])['Total Crime'].sum().reset_index()
        sorted_data = sorted_data.sort_values(by="Total Crime", ascending=(top_bottom == 'bottom'))
        
        # Clean the 'Total Crime' column
        df_punjab['Total Crime'] = df_punjab['Total Crime'].replace({'-': '0', ',': ''})  # Replace hyphens with '0' and remove commas
        df_punjab['Total Crime'] = pd.to_numeric(df_punjab['Total Crime'], errors='coerce')  # Convert to numeric, invalid values will be NaN


        # **High Alert Logic**: Check for districts exceeding the threshold
        high_alert_districts = sorted_data[sorted_data['Total Crime'] > HIGH_ALERT_THRESHOLD]

        if not top_bottom_n.empty:
            visualize_crime_type_data(top_bottom_n)

        # Show high alert notifications
        if not high_alert_districts.empty:
            alert_message = (
                "High Alert Districts:\n" +
                "\n".join(
                    f"\n{row['district']}:{row['Crime Type']}: {row['Total Crime']} crimes \n"
                    for _, row in high_alert_districts.iterrows()
                )
            )
            messagebox.showwarning("High Alert!", alert_message)
        else:
            messagebox.showinfo("All Clear", "No districts require high alert.")
        

    # Function to visualize the crime type data (graph generation)
    def visualize_crime_type_data(data):
        # Create a Plotly bar plot for Crime Type Analysis
        fig = px.bar(data, x='district', y='Total Crime', color='district', title="Crime Type Analysis - Punjab",labels={"district": "District", "Total Crime": "Number of Incidents", "Crime Type": "Type of Crime"})
        
        # Save the plot as an HTML file
        fig.write_html('punjab_citywise_crimes.html')
        webbrowser.open("punjab_citywise_crimes.html")

    # Creating Crime Type Combobox (dynamically filled from dataset)

    tk.Label(frame, text="Punjab Crime Analysis", font=("Arial", 50, "bold", "italic")).place(relx=0.5, rely=0.1, anchor="center")
    tk.Label(frame, text="SELECT CRIME :", font=("Arial", 20, "bold", 'italic')).place(relx=0.3, rely=0.3, anchor="center")
    crime_type_var = tk.StringVar()
    crime_types = df_punjab['Crime Type'].unique().tolist()  # Get unique Crime Types from the dataset
    crime_type_dropdown = ttk.Combobox(frame, textvariable=crime_type_var, values=crime_types, state='readonly', font=("Arial", 14, "bold"))
    crime_type_dropdown.bind("<<ComboboxSelected>>", update_years)  # Update years when crime type changes
    crime_type_dropdown.place(relx=0.3, rely=0.4, anchor="center")

    # Creating Year Combobox (dynamically filled based on Crime Type)
    tk.Label(frame, text="SELECT YEAR :", font=("Arial", 20, "bold", 'italic')).place(relx=0.6, rely=0.3, anchor="center")
    year_var = tk.StringVar()
    year_dropdown = ttk.Combobox(frame, textvariable=year_var, state='readonly', font=("Arial", 14, "bold"))
    year_dropdown.place(relx=0.6, rely=0.4, anchor="center")

    # Creating Top/Bottom Combobox
    tk.Label(frame, text="Top/Bottom:", font=("Arial", 20, "bold", 'italic')).place(relx=0.6, rely=0.5, anchor="center")
    top_bottom_var = tk.StringVar(value='top')
    top_bottom_dropdown = ttk.Combobox(frame, textvariable=top_bottom_var, values=['top', 'bottom'], state='readonly', font=("Arial", 14, "bold"))
    top_bottom_dropdown.place(relx=0.6, rely=0.6, anchor="center")

    # Creating Number Combobox (1-7)
    tk.Label(frame, text="Number (1-7):", font=("Arial", 20, "bold", 'italic')).place(relx=0.3, rely=0.5, anchor="center")
    number_var = tk.IntVar(value=1)
    number_dropdown = ttk.Combobox(frame, textvariable=number_var, values=[i for i in range(1, 8)], state='readonly', font=("Arial", 14, "bold"))
    number_dropdown.place(relx=0.3, rely=0.6, anchor="center")

    
    submit_button = tk.Button(frame, text="LET'S VISUALIZE !!", command=submit_inputs, bg='green', fg='white', width=30, height=3, font=("Tegami Type", 10, "bold"))
    submit_button.place(relx=0.5, rely=0.75, anchor="center")

    #Back button to return to Punjab analysis menu
    tk.Button(frame, text="BACK", command=lambda:show_frame("analysis_menu"), width=30, height=2, bg="black", fg="white", font=("Tegami Type", 10,'bold')).place(relx=0.5, rely=0.85, anchor="center")

    return frame


# Create all frames 
create_main_menu()
create_analysis_menu()
punjab_analysis_page()
create_pakistan_analysis()
create_pakistan_citywise_analysis()
create_kpk_analysis()


frames["pakistan_analysis"] = create_pakistan_analysis()
frames["visualize_all_reported"] = create_visualize_all_reported()
frames["visualize_crime_type"] = create_visualize_crime_type()
frames['kpk_crime_type_analysis']=create_kpk_crime_type_analysis()
frames['kpk_district_analysis']=create_kpk_district_analysis()

frames["punjab_crime_type_analysis"] = punjab_crime_type_analysis()
frames["punjab_district_crime_analysis"] = punjab_district_crime_analysis() 


# Show the main menu frame initially
show_frame("main_menu")

root.mainloop()
```
